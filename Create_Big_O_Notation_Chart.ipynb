```python
def linear_search(lst,target):
    for i in range(len(lst)):
        if(lst[i]== target):
            return i
    return -1
lst=np.linspace(10, 10_000, 100, dtype=int)
target=2
print(linear_search(lst,target))
if __name__ == '__main__':
    import timeit
    print(timeit.timeit("linear_search('lst','target')", setup="from __main__ import linear_search",number=1))
```
