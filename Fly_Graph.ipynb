```python
import networkx as nx
df = nx.from_pandas_edgelist(data, source='Origin', target='Dest', edge_attr=True)
```

```python
df.nodes()
```

```python
df.edges()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(12,8))
nx.draw_networkx(df, with_labels=True)
```

```python
shortest_path_distance = nx.dijkstra_path(df, source='AMA', target='PBI', weight='Distance')
shortest_path_distance
```

```python
# Initiating an empty Graph object 
P = nx.Graph() # create an empty object
# You can add nodes using add_nodes_from()
P.add_nodes_from(['A','B','C','D', 'E'])
# Use add_edges_from to add pairwise relationships
P.add_edges_from ([('B','C'), ('A','C'), ('B','D'), ('D','A'), ('D','E'), ('B','E')])
```
