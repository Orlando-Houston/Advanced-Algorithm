```python
def factorial(n):
    res=1
    for i in range (n):
        res *= i+1
```

```python

n=10000
times =[]
inputs=[]
for i in range(1000,n,50):
    start = time.process_time()
    factorial(i)
    end =time.process_time()
    times +=[end-start]
    inputs +=[i]
    
    
plt.figure(figsize=(9,3))
plt.xlabel("n")
plt.ylabel('t')

plt.scatter(inputs,times,marker=".")

# fit the data with a linear function
params, cov = optimization.curve_fit(linear_func,inputs,times)
linear_fit =[]

for x in inputs:
    linear_fit += [linear_func(x,params[0],params[1])]
    
plt.plot(inputs,linear_fit,color='red');
```

```python
def On2_bubble_sort(n):
    for i in range(len(n)-1):
        for j in range(len(n)-1-i):
            if n[j] > n[j+1]:
                n[j], n[j+1] = n[j+1], n[j]
    return n
```

## Logisitics Growth Model
* Logarithmic growth curves increase quickly in the beginning, but the gains decrease and become more difficult as time goes on.
* A much less common model for growth is logarithmic change.  The logarithm is the mathematical inverse of the exponential, so while exponential growth starts slowly and then speeds up faster and faster, logarithm growth starts fast and then gets slower and slower.

```python
def Ologn_binary_search(n,target):
    first = 0
    last = len(n) - 1
    is_found = False
    while first <= last and not is_found:
        mid = (first + last)//2
        if n[mid] == target:
            is_found = True
        else:
            if target < mid:
                last = mid - 1
            else:
                first = mid + 1
```

## What is Exponential Time Complexity?
* When the computation time of a code increases as function of X^N, N being the size of input. Such a code has Polynomial Time Complexity.

* For example, following recursive code to find Nth fibonacci number has Time Complexity as O(2^N) 

```python
def exponential(x, a, k, b):
    return a*np.exp(x*k) + b
```

```python
def F(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return F(n-1) + F(n-2) # For every call to F, we make 2 more calls to F itself
```

```python
import time
import scipy.optimize as optimization
n=50
times = []
inputs= []

for i in range(1, n, 5):
    
    start = time.process_time()
    F(i)
   
    end =time.process_time()
    times += [end-start]
    inputs += [n]
    
    
plt.figure(figsize=(9,3))
plt.scatter(inputs, times, marker='^')

plt.xlabel("n")
plt.ylabel('t')



# fit the data with a quadratic  function
params, cov = optimization.curve_fit(exponential,inputs,times)
F_fit =[]
for x in inputs:
    F_fit += [exponential(x, params[0], params[1], params[2])]
    
plt.plot(inputs, F_fit, color='g');
```

## How to Perform Simple Linear Regression in Python (Step-by-Step)

```python
import pandas as pd

#create dataset
df = pd.DataFrame({'hours': [1, 2, 4, 5, 5, 6, 6, 7, 8, 10, 11, 11, 12, 12, 14],
                   'score': [64, 66, 76, 73, 74, 81, 83, 82, 80, 88, 84, 82, 91, 93, 89]})
      

#view first six rows of dataset
df[0:6]
```

```python
import matplotlib.pyplot as plt

plt.scatter(df.hours, df.score)
plt.title('Hours studied vs. Exam Score')
plt.xlabel('Hours')
plt.ylabel('Score')
plt.show()
```

```python
df.boxplot(column=['score'])
```

```python
import statsmodels.api as sm

#define response variable
y = df['score']

#define explanatory variable
x = df[['hours']]

#add constant to predictor variables
x = sm.add_constant(x)

#fit linear regression model
model = sm.OLS(y, x).fit()

#view model summary
print(model.summary())
```

```python
#define figure size
fig = plt.figure(figsize=(12,8))

#produce residual plots
fig = sm.graphics.plot_regress_exog(model, 'hours', fig=fig)
```

```python
#define residuals
res = model.resid

#create Q-Q plot
fig = sm.qqplot(res, fit=True, line="45")
plt.show() 
```

## How to Perform Quantile Regression in Python

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
```

```python
#make this example reproducible
np.random.seed(0)

#create dataset
obs = 100

hours = np.random.uniform(1, 10, obs)
score = 60 + 2*hours + np.random.normal(loc=0, scale=.45*hours, size=100)

df = pd.DataFrame({'hours': hours, 'score': score})

#view first five rows
df.head()
```

```python
#fit the model
model = smf.quantreg('score ~ hours', df).fit(q=0.9)

#view model summary
print(model.summary())
```

```python
#define figure and axis
fig, ax = plt.subplots(figsize=(8, 6))

#get y values
get_y = lambda a, b: a + b * hours
y = get_y(model.params['Intercept'], model.params['hours'])

#plot data points with quantile regression equation overlaid
ax.plot(hours, y, color='black')
ax.scatter(hours, score, alpha=.3)
ax.set_xlabel('Hours Studied', fontsize=14)
ax.set_ylabel('Exam Score', fontsize=14)
```

```python
# General Functions
def func_exp(x, a, b, c):
    """Return values from a general exponential function."""
    return a * np.exp(b * x) + c


def func_log(x, a, b, c):
    """Return values from a general log function."""
    return a * np.log(b * x) + c


# Helper
def generate_data(func, *args, jitter=0):
    """Return a tuple of arrays with random data along a general function."""
    xs = np.linspace(1, 5, 50)
    ys = func(xs, *args)
    noise = jitter * np.random.normal(size=len(xs)) + jitter
    xs = xs.reshape(-1, 1)                                  # xs[:, np.newaxis]
    ys = (ys + noise).reshape(-1, 1)
    return xs, ys
```

```python
Relationship |  Example   |     General Eqn.     |  Altered Var.  |        Linearized Eqn.  
-------------|------------|----------------------|----------------|------------------------------------------
Linear       | x          | y =     B * x    + C | -              |        y =   C    + B * x
Logarithmic  | log(x)     | y = A * log(B*x) + C | log(x)         |        y =   C    + A * (log(B) + log(x))
Exponential  | 2**x, e**x | y = A * exp(B*x) + C | log(y)         | log(y-C) = log(A) + B * x
Power        | x**2       | y =     B * x**N + C | log(x), log(y) | log(y-C) = log(B) + N * log(x)
```
