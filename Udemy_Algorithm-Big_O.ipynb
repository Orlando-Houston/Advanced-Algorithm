```python
# Big O (n)
def print_items(n):
    for i in range(n):
        print(i)
        
print_items(10)
```

```python
def print_items_d(n):
    for i in range(n):
        print(i)
        
    for j in range(n):
        print(j)
        
print_items_d(10)
```

```python
# O(n^2) nested loop  100 items   n*n= n^2  =  O(n^2)
```

```python
def print_items_d(n):
    for i in range(n):
        for j in range(n):
            print(i,j)
        
print_items_d(10)
```

```python
# n*n*n=    O(n^3) 1000    = cube
```

```python
def print_items_d(n):
    for i in range(n):
        for j in range(n):
            for k in range(n):
                print(i,j,k)
        
print_items_d(10)
```

```python
# Drop Non-Dominants  O(n^2)
def print_items_d(n):
    for i in range(n):
        for j in range(n):
            print(i,j)
    for k in range(n):     # O(n)
        print(k)
        
print_items_d(10)
```

```python
#  O (a   +b)
```

```python
def print_items_dif(a,b):
    for i in range(a):           # O(a)
        print(i)
        
    for j in range(b):          # O(b)
        print(j)
        
print_items_dif(a,b)
```

```python
# O(a*b)
def print_items_dif(a,b):
    for i in range(a):           # O(a)
        
        for j in range(b):          # O(b)
            print(i,j)
        
print_items_dif(a,b)
```
