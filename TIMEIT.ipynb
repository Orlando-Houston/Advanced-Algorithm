```python
def factorial(n):
    res=1
    for i in range (n):
        res *= i+1
%timeit factorial(50)
```

```python
def fact(n):
    product = 1
    for i in range(n):
        product = product * (i+1)
    return product

#print (fact(5))
%timeit fact(50)
```

```python
# recursion
def fact2(n):
    if n == 0:
        return 1
    else:
        return n * fact2(n-1)

#print (fact2(5))
%timeit fact2(50)
```

```python
def lin_search(n,target):
    for i in range(len(n)):
        if(n[i]== target):
            return i
    return -1
n=np.linspace(1, 10_000_00, 1, dtype=int)
target=2
print(lin_search(n,target))
%timeit lin_search(n,target)
```

```python
import matplotlib.pyplot as plt
target = range(1, 31, 10)
y_1 = []
y_2 = []


for n in target:
    n= list(range(0, n))
    y_1.append(timeit.timeit(lambda: lin_search(n, 1000)))
    y_2.append(timeit.timeit(lambda: lin_search(target, 0)))
    
%matplotlib notebook

import matplotlib.pyplot as plt
plt.plot(target,y_1,'r') 
plt.plot(target,y_2,'g') 

plt.ylim(ymin=0)
plt.show();
```
