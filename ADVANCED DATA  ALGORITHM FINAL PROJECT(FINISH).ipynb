### What is time complexity
* Time complexity is the amount of time taken by an algorithm to run, as a function of the length of the input. It measures the time taken to execute each statement of code in an algorithm.
* Time Complexity:-
* We can define time complexity as the amount of time that the algorithm needs during the execution til generating the output.
* Time complexity affected by the number of steps in the algorithm to solve the problem.
* Time complexity is represented by Big Oh asymptotic annotation which represents the worst case that can happen by this algorithm.

#### How to analyze time complexity:
* Running  time depend upon:
* Single vs multiple processor/Read-Write speed to memory/32bit vs 64-bit/ Input


### Asymptotic Analysis
* Asymptotic analysis is the computing of an algorithms running time, and there are actually a few different variations that allow us to measure different aspects of that running time:
* Big Omega: represents the lower bound.
* Big Theta: represents both the lower and upper bounds.
* Big O: represents the upper bound.

 #### Big O() notation
### Algorithm Analysis with Big-O Notation
* Big-O notation is a metrics used to find algorithm complexity. Basically, Big-O notation signifies the relationship between the input to the algorithm and the steps required to execute the algorithm. It is denoted by a big "O" followed by opening and closing parenthesis. Inside the parenthesis, the relationship between the input and the steps taken by the algorithm is presented using "n".

#### General rules
* Ignore constant :5n---->O(n)
* certain terms 'dominate' others
* O(1)<O(logn)< O(n)<O(nlogn)<O(n2)<O(2Uzeri n)<O(n!)

* Growth Types
* O(1): constant time
* O(Log n): logarithmic time
* O(n): square root time
* O(n): linear time
* O(n Log n): linearithmic time
* O(n*n): quadratic time (squared)
* O(n^2): polynomial time
* O(2^n): exponential time
* O(n!): factorial time

####  The runtime of the factorial operation written with different algorithm

```python
def fact(n):
    product = 1
    for i in range(n):#--------------------->O(N)
        product = product * (i+1)
    return product

#print (fact(5))
%timeit fact(50)
```

```python
def factorial(n):
    res=1
    for i in range (n):
        res *= i+1
%timeit factorial(50)
```

```python
# recursion
def fact2(n):
    if n == 0:
        return 1
    else:
        return n * fact2(n-1)

#print (fact2(5))
%timeit fact2(50)
```

```python
# Linear func
def func(x,a,b):
    return a*x+b
# quadratic func
def func1(x, a, b, c):
    return a*x**2+b*x+c
# cubic func
def func2(x, a, b, c):
    return a*x**3+b*x+c

def func3(x, a, b, c):
    return a*x**3+b*x**2+c
# Exponential funct
def exp_func(x, a, b, c):
    return a*(2**x)+ b*x +c
```

```python
def my_list_sum(n):
    result = 0
    for i in range(n):
        result += i
    return result
```

```python
import time
n=100000
times =[]
inputs=[]

for i in range(1,n,1):
    start = time.process_time()
    my_list_sum(i)
    end =time.process_time()
    times +=[end-start]
    inputs +=[i]
    
    
plt.figure(figsize=(9,3))
plt.xlabel("n")
plt.ylabel('t')

plt.scatter(inputs,times,marker="^",label ='riginal')

# fit the data with a linear function
params, cov = curve_fit(func,inputs,times)
params, cov = curve_fit(func1,inputs,times)
params, cov = curve_fit(func2,inputs,times)
params, cov = curve_fit(func3,inputs,times)

linear_fit =[]
quadratic_fit =[]
cubic_fit =[]
dontnow_fit =[]



for x in inputs:
    linear_fit += [func(x,params[0],params[1])]
    quadratic_fit += [func1(x,params[0],params[1],params[2])]
    cubic_fit += [func2(x,params[0],params[1],params[2])]
    dontnow_fit += [func3(x,params[0],params[1],params[2])]
    

plt.plot(inputs,linear_fit,color='red',label = "y=a*x+b");    
plt.plot(inputs,quadratic_fit,color='black',label = "y=a*x**2+b*x+c");
plt.plot(inputs,cubic_fit,color='green', label ="y=a*x**3+b*x+c");
plt.plot(inputs,dontnow_fit,color='yellow',label = "y=a*x**3+b*x**2+c");


plt.legend(loc='upper left',fancybox=True,shadow=True)
plt.grid(True)
plt.show()
```

```python
import numpy as np

import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import FunctionTransformer


np.random.seed(123)
```

```python
# General Functions
def func_exp(x, a, b, c):
    """Return values from a general exponential function."""
    return a * np.exp(b * x) + c


def func_log(x, a, b, c):
    """Return values from a general log function."""
    return a * np.log(b * x) + c


# Helper
def generate_data(func, *args, jitter=0):
    """Return a tuple of arrays with random data along a general function."""
    xs = np.linspace(1, 5, 50)
    ys = func(xs, *args)
    noise = jitter * np.random.normal(size=len(xs)) + jitter
    xs = xs.reshape(-1, 1)                                  # xs[:, np.newaxis]
    ys = (ys + noise).reshape(-1, 1)
    return xs, ys
```
