* sum function
* The built-in Python sum function calculates the sum of all elements of a provided Iterable. This function implements an algorithm with a O(n) runtime complexity.
* To test this, we will use the linspace method from the NumPy library to generate an iterable with 50 evenly spaced values ranging from 10 to 10,000. The graph, although not a perfectly straight plot, illustrates this.

* o bunu test etmek için NumPy kitaplığındaki linspace yöntemini kullanarak 10 ile 10.000 arasında değişen 50 eşit aralıklı değere sahip bir yinelenebilir oluşturacağız. Grafik, tamamen düz bir arsa olmasa da, bunu göstermektedir.

* Running Time Analysis
* Whether a program is useful or not depends on (i) its correctness and (ii) its efficiency. I'll talk about the correctness issue later. For now let us focus on the efficiency of a program. Efficiency typically refers to:
* the amount of time the program takes and
* the amount of memory the program uses.
* As we design algorithms and programs, we need to analyze them to estimate their efficiency. The analysis can be of two kinds:
* We implement the algorithm and "profile" it by inserting extra lines of code to tell us how much time a particular function takes or how much memory is being used by a particular data structure. Note that in this approach, we are measuring the running time of a particular program, written in a particular language, using particular compiler, on a particular machine, running a particular operating system, etc. It may be hard to generalize the results we get from such profiliing.
* We do a rough, "pen-and-paper" estimate of the time and memory it takes for our algorithm to run. This is independent of language, compiler, machine, operating system, etc. However, by necessity this analysis is rough and can only show us "trends" rather than providing specific running times. However, as we will see later, knowing these trends is often enough to discard inefficient candidate algorithms.

```python
def linear_search(A, x):
    answer = -1
    for i in range(0, len(A)):
        if A[i] == x: answer = i
    return answer
```

```python
linear_search([10, 5, 9, 9], 10)
```

```python
def better_linear_search(A, x):
    for i in range(0, len(A)):
        if A[i] == x: return i
    return -1
```

```python

better_linear_search([10, 5, 9, 9], 9)
```

```python
first7Primes = [2, 3, 5, 7, 11, 13, 17]

def linearSearch(arr, item):
    for i in arr:
        if (i == item):
            return True
    return False

print(linearSearch(first7Primes , 5)) # True
print(linearSearch(first7Primes , 10)) # False
```

```python
from math import floor

first7Primes = [2, 3, 5, 7, 11, 13, 17]

def binarySearch(arr, targetValue):
    array = sorted(arr)
    min = 0
    max = len(array) - 1

    while True:
        average = floor((min + max)/2)
        if (max < min):
            return False
        if (array[average] == targetValue):
            return True
        if (array[average] < targetValue):
            min = average + 1
        else:
            max = average - 1

print(binarySearch(first7Primes, 5)) # True
print(binarySearch(first7Primes, 10)) # False
```
