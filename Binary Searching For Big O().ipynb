```python
def binary_search_recursive(array, element, start, end):
    if start > end:
        return -1

    mid = (start + end) // 2
    if element == array[mid]:
        return mid

    if element < array[mid]:
        return binary_search_recursive(array, element, start, mid-1)
    else:
        return binary_search_recursive(array, element, mid+1, end)
    
element = 18
array = [1, 2, 5, 7, 13, 15, 16, 18, 24, 28, 29]

if __name__ == '__main__':
    import timeit
    print(timeit.timeit("binary_search_recursive("array", "element", "start", "end")", setup="from __main__ import binary_search_recursive",number=1))
```

* What is recurrence relation for binary search algorithm?
* Time complexity is O(logN)-

* Recurrence relation-> T(n)=T(n/2)+1

* Derivation->

* 1st step=> T(n)=T(n/2) + 1

* 2nd step=> T(n/2)=T(n/4) + 1 ……[ T(n/4)= T(n/2^2) ]

* 3rd step=> T(n/4)=T(n/8) + 1 ……[ T(n/8)= T(n/2^3) ]

.

.

* kth step=> T(n/2^k-1)=T(n/2^k) + 1*(k times)

* Adding all the equations we get, T(n) = T(n/2^k) + k times 1 _____eq(final)

* => n/2^k= 1 [So how many times we need to divide by 2 until we have only one element left]

* => n=2^k

* => log n=k [taken log(base 2) on both sides ]

* Put k= log n in eq(final)

* T(n) = T(1) + log n

* T(n) = 1 + log n [we know that T(1) = 1 , because it’s a base condition as we are left with only one element in the array and * that is the element to be searched so we return 1]

* T(n) = O(log n) [taking dominant polynomial, which is n here)

* This is how we got “log n” time complexity for binary search.

```python
def binary_search_r(array, low, high, x):
 
    
    if high >= low:
        mid = (high + low) // 2
 
        
        if array[mid] == x:
            return mid
 
        elif array[mid] > x:
            return binary_search_r(array, low, mid - 1, x)
 
        else:
            return binary_search_r(array, mid + 1, high, x)
 
    else:
        return -1

array = [ 2, 4, 6, 8, 20, 40, 60, 80]
x = 20

result = binary_search_r(array, 0, len(array)-1, x)
 
if result != -1:
    print("The index of the Element is", str(result))
else:
    print("This element is not present in your Array.")
```

```python
# Recursive implementation of the binary search algorithm to return
# the position of `target` in subarray nums[left…right]
def binarySearch(nums, left, right, target):
 
    # Base condition (search space is exhausted)
    if left > right:
        return -1
 
    # find the mid-value in the search space and
    # compares it with the target
 
    mid = (left + right) // 2
 
    # overflow can happen. Use below
    # mid = left + (right - left) / 2
 
    # Base condition (a target is found)
    if target == nums[mid]:
        return mid
 
    # discard all elements in the right search space,
    # including the middle element
    elif target < nums[mid]:
        return binarySearch(nums, left, mid - 1, target)
 
    # discard all elements in the left search space,
    # including the middle element
    else:
        return binarySearch(nums, mid + 1, right, target)
 
 
if __name__ == '__main__':
 
    nums = [2, 5, 6, 8, 9, 10]
    target = 5
 
    (left, right) = (0, len(nums) - 1)
    index = binarySearch(nums, left, right, target)
 
    if index != -1:
        print('Element found at index', index)
    else:
        print('Element found not in the list')
 
```

```python
def count_rotations_linear(nums):
    # If list is empty
    if len(nums) == 0:
        return -1
    
    else:
        position = 0
        
        # Loop until last position in nums
        while position <= len(nums)-2:                     

            # Success criteria: check whether the number at the current position is smaller than the one before it
            if position >= 0 and nums[position] > nums[position+1]:  
                return position + 1

            # Move to the next position
            position += 1
        
        # If list is unrotated
        return 0
```

```python
def binary_search(lo, hi, condition, nums_len):
    while lo < hi:
        # Calculating mid position
        mid = (lo + hi) // 2 
        # result of binary search
        result = condition(mid, hi, lo)
        
        if result == 'found':
            return mid + 1
        elif result == 'left':
            # Dropping right half
            hi = mid
        elif result == 'right':
            # Dropping left half
            lo = mid + 1
        elif result == 'unrotated':
            return 0
    # "empty"    
    return -1
```

```python
def count_rotations_binary(nums):
    
    def condition(mid, hi, lo):
        mid_number = nums[mid]
        
        if hi - lo == len(nums)-1 and nums[lo] < nums[hi]:
            # If list is unrotated
            return "unrotated"
        elif len(nums) == 1:
            return "unrotated"
        else:            
            # Success criteria: mid_number is greater or equal to number in lowest position and greater than the next
            if mid_number > nums[mid+1]:
                # The middle position is the answer
                return "found"
            elif mid_number > nums[lo]:
                # Answer lies in the right half
                return "right"
            elif mid_number < nums[lo]:
                # Answer lies in the left half
                return "left"
        
    return binary_search(0, len(nums)-1, condition, len(nums) // 2)
```

```python
from time import perf_counter as timer
from  collections import Counter
import math
```

```python
def evaluate_cases(count_rotations, tests, display=True):
    results = []
    i = 1
    for test in tests:
        # Timing the search opperation passed as "count_rotations"
        start = timer()
        position = count_rotations(test['input']['nums'])
        end = timer()
        
        if display:
            print("Test case #{}".format(i), "\n")
            print("Input:\n'nums': {}\n".format(test['input']['nums']))

            print("Expected position:\n{}\n".format(position))
            print("True position:\n{}\n".format(test['output']))

            print("Execution time:\n{:.2f} ms\n".format((end - start)*1e3))        
        else:
            print("Position:", position)
            print("Execution time: {:.2f} ms".format((end - start)*1e3))
        
        if position == test['output']:
            print("Test result:\nPASSED\n")
            results.append("passed")
        else:
            print("Test result:\nFAILED\n")
            results.append("failed")
        
        i += 1
        
    if display:
        print("SUMMARY:\n")
        print("TOTAL: {}".format(len(results)), 
              "PASSED: {}".format(Counter(results)['passed']), 
              "FAILED: {}\n".format(Counter(results)['failed']))
```
